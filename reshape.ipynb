```python
import numpy as np

x = np.arange(8).reshape(8, 1)

candidates = [
    (2, 4, 4),
    (2, 2, 2),
    (1, 4, 3),
    (-1, 3),
]

for shape in candidates:
    try:
        y = x.reshape(shape)
        print(f"reshape{shape} -> OK, result shape: {y.shape}")
    except Exception as e:
        print(f"reshape{shape} -> Error: {e}")
```

```python
# python
import numpy as np

x = np.arange(8)[:, None]   # shape (8, 1)
candidates = [(2, 2, 2), (8,), (1, 8), (-1, 3)]

for s in candidates:
    try:
        y = x.reshape(s)
        print("reshape", s, "-> OK, shape:", y.shape)
    except Exception as e:
        print("reshape", s, "-> Error:", e)
```

```python
# python
import numpy as np

np.random.seed(0)
A = np.random.randn(3, 3)   # shape (3,3)
v = np.random.randn(3, 1)   # shape (3,1)

# element-wise with broadcasting
C_elem = A * v

# matrix multiplication
C_dot = np.dot(A, v)

print("A.shape =", A.shape)
print("v.shape =", v.shape)
print("A * v -> shape:", C_elem.shape)
print(C_elem)
print("np.dot(A, v) -> shape:", C_dot.shape)
print(C_dot)

# quick checks
assert C_elem.shape == (3, 3)
assert C_dot.shape == (3, 1)
```
